```python
import requests 
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://news.ycombinator.com/news'
```

```python
r=requests.get(url)
```

```python
html_content=BeautifulSoup(r.text,'html.parser')
```

```python
import re
html_content
dictionary=[]
links=[]
titles=[]
points=[]
comments=[]
for items in html_content.findAll('tr',class_='athing'):
    a_tag=items.find('a',class_='storylink')
    link=a_tag.get('href') if a_tag else None
    title=a_tag.get_text(strip=True) if a_tag else None
    next_row=items.findNextSibling('tr')
    span=next_row.find('span',class_='score')
    point=span.get_text(strip=True) if span else '0 Points'
    print(link)
    print(title)
    print(point)
    item_comments = next_row.find('a', string=re.compile('\d+(&nbsp;|\s)comment(s?)'))
    item_comments = item_comments.get_text(strip=True).replace('\xa0', ' ') \
                        if item_comments else '0 comments'
    print(item_comments)
    dictionary.append({'Link':link,'Title':title,'Point':point,'Comment':item_comments})
    links.append(link)
    titles.append(title)
    points.append(points)
    comments.append(item_comments)
    
    
for items in dictionary:
    print(items)
```

```python
#Now we have list of links,points,titles and comments
#Now we are ready to put all this into dataframe

import pandas as pd
```

```python
df=pd.DataFrame({'Links':links,'Titles':titles,'Points':points,'Comments':comments})
```

```python
import os
import re;

directory ='/home/shashanksoni092/MIT/RegexPython'
principalDiagnosis=[]
dischargeDate=[]
admissionDate=[]
age=[]
date=[]
time=[]
address=[]

features=[principalDiagnosis,dischargeDate,admissionDate,age,date,time,address]

mainfields=[]
count=0;
for filename in os.listdir(directory):
    
    if(filename):
        tgz_path = os.path.join(directory,filename)
        f = open(tgz_path)
        fields=[]
        print("file:",filename)
        for l in f:
            #print("Hi")
            PrincipalDiagnosis=re.search("PRINCIPAL DIAGNOSIS:*",l)
            if(PrincipalDiagnosis):
                print("PrincipalDiagnosis:",l[20:])
                fields.append(l[20:])
                principalDiagnosis.append(l[20:])

            DischargeDate=re.search("Discharge Date:*",l)
            if(DischargeDate):
                print("Discharge Date:",l[15:])
                fields.append(l[15:])
                dischargeDate.append(l[15:])

            AdmissionDate=re.search("Admission Date:*",l)
            if(AdmissionDate):
                print("Admission Date:",l[AdmissionDate.start()+15:AdmissionDate.start()+25])
                fields.append(l[AdmissionDate.start()+15:AdmissionDate.start()+25])
                admissionDate.append(l[AdmissionDate.start()+15:AdmissionDate.start()+25])
                
            Age=re.search("..[-\s]year[-\s]old",l)
            if(Age):
                print("Age:",l[Age.start():Age.start()+2])
                fields.append(l[Age.start():Age.start()+2])
                age.append(l[Age.start():Age.start()+2])

            date_time= re.search('(\d+)[/](\d+)[/](\d{4})\s(\d+)[:]',l)
            if(date_time):
                print("Date:",l[date_time.start():date_time.start()+10])
                fields.append(l[date_time.start():date_time.start()+10])
                date.append(l[date_time.start():date_time.start()+10])
                
                print("Time:",l[date_time.end()-3:date_time.end()+8])
                fields.append(l[date_time.end()-3:date_time.end()+8])
                time.append(l[date_time.end()-3:date_time.end()+8])
                
            Address=re.search('(\w)+\sHospital',l)
            if(Address):
                print("Address:",Address.group())
                fields.append(Address.group())
                address.append(Address.group())
                break

        for i in range(len(features)):
            if(len(features[i])<count+1):
                features[i].append("null")
        print(fields) 
        mainfields.append(fields)
    else:
        print("Hello")
        continue
    count +=1    
print(mainfields)
print(len(mainfields))

for i in range(len(features)):
    print(features[i],len(features[i]))
```
